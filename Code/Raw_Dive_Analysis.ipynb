# Raw Dive Analysis (before rolling wdw)

**Imports**

```r
library(data.table)
```

**Create vector**

```r
cat('\nExtracting pressure data...\n\n')

# Get all Pressure files (containing Pressure/activity data)
files = list.files("../Data/BIOT_DGBP/BIOT_DGBP/", pattern = "1.csv", full.names = TRUE)
depths = c()
depths_mod = c()

# For each file, extract pressure data and resave as *_p.csv
for (i in 1:length(files)) {
  
  f = files[i]
  cat(sprintf("Processing file: %s\n", f))
  
  cat("Reading in file...")
  
  # Read in file and subset rows containing pressure data
  ts_data = fread(file = f)
  ts_data_d = ts_data[!is.na(Depth)]
  depths = c(depths, ts_data_d$Depth)
  depths_mod = c(depths_mod, ts_data_d$Depth_mod)
}
```

**Extract and analyse dives**

```r
threshold = 0.4
dives = depths[depths>threshold]
dives = dives[dives<9] # drop drastic anomalies
```

```r
length(dives)
```

```r
hist(dives, 1000)
```

**Stats**

```r
cat(sprintf('Max:\t%.2f\n', max(dives)))
cat(sprintf('Mean:\t%.2f\n', mean(dives)))
cat(sprintf('Sd:\t%.2f\n', sd(dives)))
```

# Zero-offset depth

```r
threshold = 0.03
dives_mod = depths_mod[depths_mod>threshold]
dives_mod = dives_mod[dives_mod<9] # drop drastic anomalies
length(dives_mod)
```
